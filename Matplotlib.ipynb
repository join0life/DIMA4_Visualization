## 범례 표시하기
- plot() 함수에 label 키워드 인자를 설정함으로써 그래프에 이름을 지정할 수 있다.
- 이렇게 지정된 이름을 legend() 함수를 통해 범례로 표시할 수 있다.
- loc 파라미터에 다음 문자열을 지정하여 범례의 위치를 정할 수 있다.
<pre>
    'best' -> 알아서 최적의 위치 선정해줌
    'upper right'
    'upper left'
    'lower left'
    'lower right'
    'right'
    'center left'
    'center right'
    'lower center'
    'upper center'
    'center'
</pre>

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(0, 3, .02)
c = np.exp(a)
d = c[::-1] # c를 뒤집은 거


plt.plot(a, c, 'r--', label = 'Model length')
plt.plot(a, d, 'b:', label = 'Data length')
plt.plot(a, c+d, 'g', label = 'Total message length')

plt.legend(loc = 'upper right', shadow=True) # 범례 우측 상단에 위치. 범례 값을 따로 주지 않으면 best가 디폴트
plt.show()
```

## 서브플롯 그리기

### subplot() 함수
- subplot() 함수를 이용해서 하나의 그림에 여러 개의 그래프를 그릴 수 있다.
    - nrow, ncols, index의 인수를 이용하여 서브플롯의 위치를 설정한다.

```python
x1 = np.linspace(0, 5, 50) #시작(0)과 끝 값(5)을 기준으로 균등 분할(50개)
x2 = np.linspace(0, 2, 50)

y1 = np.cos(2 * np.pi * x1) + np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1) #nrow =2, ncol =1, index =1(2행 1열의 첫번째)
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.xlabel('time(s)')
plt.ylabel('Dampled')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.title('2st Graph')
plt.xlabel('time(s)')
plt.ylabel('Undampled')

plt.tight_layout() # 서브플롯의 레이아웃을 자동으로 맞춰준다.
plt.show()
```

### subplots() 함수
- subplots() 함수는 Figure, Axes 객체를 반환한다.
- Axes 클래스: 공간이 확보된 이미지 영역 정보를 가지고 있다.
    - 주요인자
    >- nrows, ncols : 서브플롯의 행과 열 크기 지정
    >- sharex, sharey : boolean 값(default:False), x, y축의 값 공유 여부

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 4) #sharey = True)

fig.set_size_inches((16,4))

axes[0].scatter(np.random.random(100), np.random.random(100), c = 'cyan', edgecolor = 'red', alpha = 0.5) #alpha는 투명도 값. (0~1 사이)  
axes[1].scatter(np.random.random(100), np.random.random(100), c = 'magenta', edgecolor = 'green', alpha = 0.5)  
axes[2].scatter(np.random.random(100), np.random.random(100), c = 'yellow', edgecolor = 'blue', alpha = 0.5)  
axes[3].scatter(np.random.random(100), np.random.random(100), c = 'white', edgecolor = 'black', alpha = 0.5)

axes[0].set_title('subplot-0')
axes[1].set_title('subplot-1')
axes[2].set_title('subplot-2')
axes[3].set_title('subplot-3')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, sharey = True)

fig.set_size_inches((8,8))

axes[0][0].scatter(np.random.random(100), np.random.random(100), c = 'cyan', edgecolor = 'red', alpha = 0.5) #alpha는 투명도 값. (0~1 사이)  
axes[0][1].scatter(np.random.random(100), np.random.random(100), c = 'magenta', edgecolor = 'green', alpha = 0.5)  
axes[1][0].scatter(np.random.random(100), np.random.random(100), c = 'yellow', edgecolor = 'blue', alpha = 0.5)  
axes[1][1].scatter(np.random.random(100), np.random.random(100), c = 'white', edgecolor = 'black', alpha = 0.5)

axes[0][0].set_title('subplot-0')
axes[0][1].set_title('subplot-1')
axes[1][0].set_title('subplot-2')
axes[1][1].set_title('subplot-3')

plt.show()
```

## 수평 막대 그래프

```python
x = np.arange(0, 20)
y = x*2

y_ticks = np.arange(len(x)) # x축의 눈금을 표시할 index 값
# 수평 막대 그래프를 그린다고 해서 x와 y 값의 위치가 바뀌는 건 아니다.
plt.barh(x, y, label = 'Blue bar', color = 'blue', edgecolor = 'yellow', linewidth = 3)
plt.ylabel('bar number')
plt.xlabel('bar height')
plt.title('H-Bar Chart Example')
plt.yticks(y_ticks, y_ticks+1)
plt.legend()
plt.show()
```

```python
label = ['Seoul', 'Busan', 'Incheon','Gwangju','Jeju']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x_ticks_index = np.arange(len(men_means))
width = 0.35 # 막대 그래프 하나의 폭의 크기
plt.bar(x_ticks_index-width/2, men_means, width = width, color ='dodgerblue', label = 'Men') # 눈금선을 기준으로 왼쪽 막대 그래프
plt.bar(x_ticks_index+width/2, women_means, width = width, color = 'violet', label ='Women') #눈금선을 기준으로 오른쪽 막대 그래프

plt.xlabel('City')
plt.ylabel('Score')
plt.title('Score by group and gender')
plt.xticks(x_ticks_index)
plt.legend()
plt.show()
```

# 산점도 그리기
- 산점도(scatter plot)는 직교좌표계를 이용해 두 변수 간의 상관관계를 확인하고자 할 때 사용하는 플롯이다.

```python
x1 = [2, 3, 4]
y1 = [5, 5, 5]

x2 = [1, 2, 3, 4, 5]
y2 = [2, 3, 2, 3, 5]

y3 = [6, 8, 7, 8, 7]

plt.scatter(x1, y1)
plt.scatter(x2, y2, marker = 'v', color = 'green')
plt.scatter(x2, y3, marker = '^', color = 'red')
plt.title('Scatter Plot Example')
plt.show()
```

```python
# 랜덤한 값의 산점도 
for color in ['blue', 'orange', 'green']:
    n = 50
    x, y = np.random.rand(2, n) # [0,1) 사이의 난수 2 X 50 모양으로 생성
    scale = 190.0 * np.random.rand(n) + 10 # [10,200) 사이의 난수 50개 발생
    plt.scatter(x, y, c=color, s = scale, label = color, alpha = 0.3, edgecolors = 'none')

plt.legend(loc = (1.04, 0.7))
plt.grid(True)
plt.show()
```

# 히스토그램 그리기
- 자료의 분포를 몇 개의 구간으로 나눈 후 각 구간에 해당하는 데이터의 빈도수를 표로 나타낸 것을 도수분포표(frequency table)라고 한다.
- 히스토그램은 도수분포표에서 각 구간별 데이터의 빈도수를 막대 그래프로 나타낸 것이다.

```python
y = np.random.randn(1000) # 평균이 0, 표준편차가 1인 표준정규분포 난수 1000개 생성
print(np.mean(y), np.std(y)) # mean -> 평균, std -> 표준편차 구하기
x = np.arange(len(y))
plt.bar(x,y)
plt.title('Raw Data <Bar Chart>')
plt.show()

plt.hist(y, bins = 50) # bins : 데이터를 분할할 구간 수
plt.title('Histogram')
plt.show()

plt.hist(y, cumulative = True, bins = 20)
plt.title('Cumulative Histogram') # 누적 히스토그램
plt.show()
```

# 원 그래프 그리기 

```python
import matplotlib.pyplot as plt

label = ['Python', 'Java', 'C']
# autopct: 각 조각 위에 비율(%)을 표시하는 형식 문자열을 지정한다.
plt.pie([0.3, 0.35, 0.4], labels = label, autopct = '%.1f%%')
plt.title('Popular Language Ranking')
plt.legend(loc = (1, 0.3))
plt.show()
```

```python
data = [0.3, 0.35, 0.4]
label = ['Python','Java','C']
colors = ['coral','dodgerblue','yellowgreen']

plt.pie(
    data, #조각의 크기를 결정하는 값
    labels = label, # 조각의 이름 표시
    explode = (0.1, 0, 0), # 조각을 중심으로부터 얼마나 떨어지게 할 것인지를 설정한다. (0.1: 반지름의 10% 만큼 벗어남)
    colors = colors, # 조각의 색깔
    autopct = '%.2f%%', # 비율을 표시하는 형식 문자열
    shadow = True, # 그림자 표시 여부
    startangle = 90, # 조각의 시작 부분을 각도로 표시. 해당 각도에서부터 반시계 방향으로 회전하면서 조각을 표시 (기본값: 0, 3시 방향)
    counterclock = False, # 반시계방향으로 순서대로 표시할지 결정(기본값:True)
    rotatelabels = True # 조각의 이름을 회전한다. (기본값:False)
)
plt.title('Popular Language Ranking')
plt.show()
```

# 박스플롯 그리기
- 박스플롯은 데이터 집합의 범위와 중앙값 등 통계 수치 값을 확인하고자 하는 목적으로 사용한다.
- 이상치(outlier)가 있는지 검사하기 수월하다.
- 최소값 : 제 1사분위에서 1.5IQR(Interquartile Range:사분위수범위, Q3-Q1)을 뺀 위치
- 제 1사분위(Q1): 중앙값 기준으로 하위 50% 중의 중앙값, 전체 데이터 중 하위 25%에 해당하는 값
- 제 2사분면(Q2): 50%의 위치에 중앙값(median)을 의미
- 제 3사분면(Q3): 중앙값 기준으로 상위 50% 중의 중앙값, 전체 데이터 중 상위 25%에 해당하는 값
- 최대값: 제 3사분위에서 1.5IQR을 더한 위치
- 최대값, 최소값 범위를 벗어나는 값들을 이상치 데이터로 간주한다.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000) # 평균 0, 표준편차 2.0, 1000개
data_b = np.random.normal(-3.0, 1.5, 500)
data_c = np.random.normal(1.2, 1.5, 1500)

plt.boxplot([data_a, data_b, data_c])
plt.ylim(-10.0, 10.0)
plt.xlabel('Data Type')
plt.ylabel('Value')
plt.show()
```

```python
data_a = np.random.normal(0, 2.0, 1000)
print(f'min: {np.min(data_a):.2f}, median: {np.median(data_a):.2f}, max: {np.max(data_a):.2f}')
quantile = np.quantile(data_a, [0, 0.25, 0.5, 0.75, 1]) # 전체 데이터의 최하위값, 0.25값, 중간값, 0.75값, 최상위값
print(quantile)
IQR = quantile[3] - quantile[1] # 3사분위 값 - 1사분위 값 
print('IQR: ', IQR)
print('MIN: ', quantile[1] - (1.5 * IQR))
print('MAX: ', quantile[3] + (1.5 * IQR))
```

```python
# 이상치 데이터 확인

np.random.seed(0)

spread = np.random.rand(50) * 100 # [0, 100) 데이터 50개
center = np.ones(25) * 50 # 50(중앙값) 25개
flier_high = np.random.rand(10) * 100 + 100 #[100, 200) 데이터 10개
flier_low = np.random.rand(10) * -100 # (-100, 0) 데이터 10개

data = np.concatenate([spread, center, flier_high, flier_low])
print(data.shape)

plt.boxplot(data, 1, 'rx')
plt.grid(True)
plt.title('Box Plot')
plt.show()
```

```python
import random

random.seed(10)
x_invest = list(range(5, 46, 5))
y_tv = [random.randint(1, 64) for i in range(9)]
y_tv.sort()
y_google_ads = [random.randint(1, 74) for i in range(9)]
y_google_ads.sort()

plt.plot(x_invest, y_tv, 'bs-', label='TV') #s : square
plt.plot(x_invest, y_google_ads, 'ro--', label = 'Google ads')
plt.title('광고 매체별 상품 판매 효과 비교')
plt.xlabel('광고 투자 금액', labelpad = 10)
plt.ylabel('매출', labelpad = 10)
plt.legend()
plt.show()
```
